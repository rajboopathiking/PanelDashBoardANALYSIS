```python
flight_routes = df.Route.str.split("→",expand=True)
flight_routes.head()
```

```python
flight_routes.columns = ["Route1","Route2","Route3","Route4","Route5","Route6"]
```

```python
df = pd.concat([df,flight_routes],axis=1)
df.head()
```

```python
df["Dep_Time"] = pd.to_datetime(df["Dep_Time"])
df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"])
df["Arrival_Time "] = pd.to_datetime(df["Arrival_Time"])
df["Year_of_Journey"] = pd.to_datetime(df["Date_of_Journey"]).dt.year
df["Month_of_Journey"] = pd.to_datetime(df["Date_of_Journey"]).dt.month
```

```python
for i in flight_routes.columns:
    df[i] = df[i].replace({None:"Travel-ended"})
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```
